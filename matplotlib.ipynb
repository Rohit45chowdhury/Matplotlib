```python
import matplotlib.pyplot as plt
```

```python
x=['Mon','Tues','wed','thru','Fri']
y=[10,15,7,20,12]

plt.plot(x,y)
plt.title('Bakery sales this week!')
plt.xlabel('Days of week')
plt.ylabel('sales per day')
plt.show()
```

```python
month=[1,2,3,4]
sales=[1000,1500,1200,1800]

plt.plot(month,sales,color='red',linestyle='--',linewidth=2,marker=0,label='2025 sales data')
plt.xlabel('Month')
plt.ylabel('sales per month')
plt.title('monthly sale data report')
plt.legend(loc='upper left')
plt.grid(color='gray',linestyle=':',linewidth=1)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

1. Barplot

```python
product=['A','B','C','D']
sales=[1000,1500,1200,1800]

plt.bar(product,sales,color='orange',label='sale 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('product sale comparision')
plt.legend()
plt.show()
```

2. Piechart

```python
regions=['North','South','West','East']
revenue=[2000,4000,7000,12000]

plt.pie(revenue,labels=regions,autopct="%1.1f%%",colors=['blue','gold','green','red'])
plt.title('Revenue contribution by Region')
plt.show()
```

3. Histogram

```python

scores = [22,31,44,56,76,87,89,44,90,46,73]

plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.title('Score Distribution of Students')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.show()
```

4. Scatterplot

```python
hours_study=[1,2,3,4,5,6,7,8]
exam_score=[50,55,60,70,75,80,85,90]

plt.scatter(hours_study,exam_score,color='green',marker='o',label='student Data')
plt.xlabel('hours_study')
plt.ylabel('exam_score')
plt.grid(True)
plt.legend()
plt.show()
```

5. Subplot

```python
import matplotlib.pyplot as plt

scores = [22,31,44,56,76,87,89,44,90,46,73]

# 1st subplot → Histogram
plt.subplot(1, 2, 1)   # 1 row, 2 columns, position 1
plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.title('Histogram')

# 2nd subplot → Line plot
plt.subplot(1, 2, 2)   # 1 row, 2 columns, position 2
plt.plot(scores, marker='o', color='green')
plt.title('Line Plot')

plt.suptitle('Student Score Analysis')  # Overall title
plt.tight_layout()
plt.show()
```

How to save Figure

```python
x=[1,2,3,4]
y=[10,12,20,25]

plt.plot(x,y,color='blue',marker='o')
plt.title('simple line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.savefig('line_plot.png',dpi=300,bbox_inches='tight')
plt.show()
```
